## Test adding faux date for mongoDB filtering

```python
df_test['DATE_FILTER'] = ''
```

```python
for index, row in df_test.iterrows():
    date = row['DATE']
    faux_date = date + '-2008'
    df_test.loc[index, 'DATE_FILTER'] = faux_date
```
